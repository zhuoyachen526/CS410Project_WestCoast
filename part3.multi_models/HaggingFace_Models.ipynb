```python
from happytransformer import HappyTextClassification
```

```python
#this one can only do 2 categories
distilbert = HappyTextClassification(model_type="DISTILBERT", model_name="distilbert-base-uncased-finetuned-sst-2-english", num_labels=2)
result = distilbert.classify_text("I love you")
print(result.label)
print(result.score)
result = distilbert.classify_text("I am having my breakfast")
print(result.label)
print(result.score)
result = distilbert.classify_text("I hate you")
print(result.label)
print(result.score)
```
